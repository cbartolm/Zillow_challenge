# Web Scraping for Extra information 


```python
import pandas as pd
```

```python
data_set = pd.read_csv("Data/cleaned_data_R.csv")
```

```python
# data_set_copy = data_set.loc[1:3,:].copy(deep=True)
```

```python
data_set_copy = data_set.copy(deep=True)
```

```python
data_set_copy = data_set_copy.iloc[0:2,:]
```

```python
# data_set_copy
```

```python
new_df = pd.DataFrame()
for i, row in data_set_copy.iterrows():
    try:
        url_string = 'https://www.ffiec.gov/census/report.aspx?\
year=2016&county={county}&tract={number}&state=06&report=demographic'\
        .format(county= str(row.tract_county)[1:], number = row.tract_number)
        aaron_genius = pd.read_html(url_string, attrs = {"id":"Report1_dgReportDemographic"}, header=0)[0]
#         display(aaron_genius)
        new_df = pd.concat([new_df, aaron_genius])
    except:
        print(i)
        new_df = pd.concat([new_df, aaron_genius])

```

```python
new_df = new_df.set_index(data_set_copy.index)
final_df = pd.concat([data_set_copy, new_df], axis=1)
```
